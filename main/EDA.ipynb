```python
abilities_set = set()

for ability_list in df_pokemon["abilities"]:
    abilities_set.update(ability_list)

print(sorted(abilities_set))
```

```python
def get_ability_description(ability_name, df_ability_info):
    match = df_ability_info[df_ability_info["Name"].str.lower() == ability_name.lower()]
    if not match.empty:
        print(f"{ability_name}:")
        print(f"Type: {match.iloc[0]['Type']}")
        print(f"Category: {match.iloc[0]['Category']}")
        print(f"Effect: {match.iloc[0]['Effect']}")
        print(f"Power: {match.iloc[0]['Power']}")
        print(f"Accuracy: {match.iloc[0]['Acc']}")
        print(f"Power Point cost: {match.iloc[0]['PP']}")
        print(f"Second effect probability: {match.iloc[0]['Prob.(%)']}%")

    else:
        print(f"No info found for '{ability_name}'.")

get_ability_description("Thunder", df_moves)

```
